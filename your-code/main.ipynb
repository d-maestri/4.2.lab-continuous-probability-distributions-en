## Uniform Distribution

discussed in class:

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
def generate_random_numbers(bottom, ceiling, count):
    return np.random.uniform(low=bottom, high=ceiling, size=count)
```

```python
# your code here
dataset1 = generate_random_numbers(bottom=10, ceiling=15, count=100)
dataset1.shape
dataset1
```

```python
# your code here
dataset2 = generate_random_numbers(bottom=10, ceiling=60, count=1000)
dataset2.shape
dataset2
```

How are the two distributions different?

The second dataset (with bottom=10, ceiling=60, count=1,000) appears to be a better visual approximation of a continuous uniform distribution than the first dataset (bottom=10, ceiling=15, count=100). The larger sample size allows its observed distribution to more accurately represent the theoretical uniform distribution, and the wider range makes this visual representation clearer with the given binning of 10.

```python
# your answer below
plt.figure(figsize=(14, 7))
# Plot for dataset1
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.histplot(dataset1, bins=10, kde=False) # kde=False as requested for count on y-axis
plt.title('Uniform Distribution: Bottom=10, Ceiling=15, Count=100')
plt.xlabel('Value')
plt.ylabel('Count')
plt.xlim(9, 16) # A bit wider than 10-15 to show boundaries clearly

#Plot for dataset2
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.histplot(dataset2, bins=10, kde=False)
plt.title('Uniform Distribution: Bottom=10, Ceiling=60, Count=1,000')
plt.xlabel('Value')
plt.ylabel('Count')
plt.xlim(5, 65) # A bit wider than 10-60 to show boundaries clearly

plt.tight_layout() # Adjusts plot parameters for a tight layout
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
from numpy import random
def expo_numbers(mean_val, size):
    return random.exponential(scale=mean_val, size=size)

expo_10 = expo_numbers(mean_val=10, size=10)
print(expo_10)

expo_100 = expo_numbers(mean_val=100, size=100)
print(expo_100)

#plot
plt.subplot(1, 2, 1) #1 row, 2 columns, 1st plot
plt.hist(expo_10, bins=100)
plt.title("exponential distribution size 10")

plt.subplot(1, 2, 2)
plt.hist(expo_100, bins=100)
plt.title("exponential distribution size 100")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well.
